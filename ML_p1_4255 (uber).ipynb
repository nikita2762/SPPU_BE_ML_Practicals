```python
df.pickup_datetime = pd.to_datetime(df.pickup_datetime)
df['hour']=df.pickup_datetime.dt.hour
df['day']=df["pickup_datetime"].dt.day
df['month']=df.pickup_datetime.dt.month
df['year']=df.pickup_datetime.dt.dayofweek
df['dayofweek']=df.pickup_datetime.dt.dayofweek
```

```python
def remove_outlier(df1,col):
    Q1=df1[col].quantile(0.25)
    Q3=df1[col].quantile(0.75)
    
    IQR = Q3-Q1
    lower_whisker=Q1-1.5*IQR
    upper_whisker=Q3+1.5*IQR
    
    df[col]=np.clip(df[col],lower_whisker,upper_whisker)
    
    return df1
```

```python
def treat_outliers_all(df,col_list):
    for c in ["fare_amount","pickup_longitude","pickup_latitude",    "dropoff_longitude",    "dropoff_latitude",     "passenger_count"]:
        df1=remove_outlier(df,c)
    return df1
```

```python
df=treat_outliers_all(df,df.iloc[:,0::])
```

```python
df.plot(kind="box",subplots=True,layout=(7,2),figsize=(15,20))
```

```python
from sklearn.metrics.pairwise import haversine_distances
from math import radians
travel_dist=[]
for pos in range(len(df['pickup_longitude'])):
 long1,lati1,long2,lati2=[df['pickup_longitude'][pos],df['pickup_latitude'][pos],df['dropoff_longitude'][pos],df['dropoff_latitude'][pos]]
 
 loc1=(lati1,long1)
 loc2=(lati2,long2)
 
 loc1_in_radians = [radians(_) for _ in loc1]
 loc2_in_radians = [radians(_) for _ in loc2]
 result = haversine_distances([loc1_in_radians, loc2_in_radians]) * 6371.0
 travel_dist.append(result[0][1])
 
print(travel_dist)
df['dist_travel_km']=travel_dist
df.head()
```

```python
print("Data type of 'dist_travel_km':", df['dist_travel_km'].dtype)
print("Number of missing values in 'dist_travel_km':", df['dist_travel_km'].isnull().sum())
```

```python
df=df.loc[(df.dist_travel_km >=1) | (df.dist_travel_km <=130)]
print("Remaining observations in the dataset :",df.shape)
```

```python
incorrect_coordinates=df.loc[(df.pickup_latitude >90) | (df.pickup_latitude <-90) | (df.dropoff_latitude >90) | (df.dropoff_longitude > 180) | (df.pickup_longitude <-180) | (df.dropoff_longitude > 90) | (df.dropoff_longitude <-90)]
```

```python
df.drop(incorrect_coordinates,inplace=True,errors="ignore")
```
