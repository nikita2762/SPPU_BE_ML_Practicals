```python
import pandas as pd
import numpy as np #provides support for large, multi-dimensional arrays and matrices
from sklearn.model_selection import train_test_split  #easy splitting of data into training and testing sets
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier 
from sklearn import metrics
```

```python
df['Prediction'] = df['Prediction'].replace({0:'Not spam', 1:'Spam'})
```

```python
#KNN
X = df.drop(columns='Prediction',axis = 1)
Y = df['Prediction']
```

```python
 x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2)
```

```python
 KN = KNeighborsClassifier
knn = KN(n_neighbors=7)
knn.fit(x_train, y_train)
y_pred = knn.predict(x_test)
```

```python
 print("Prediction: \n")
print(y_pred)
```

```python
 M = metrics.accuracy_score(y_test,y_pred)
print("KNN accuracy: ", M)
```

```python
C = metrics.confusion_matrix(y_test,y_pred)
print("Confusion matrix: ", C)
```

```python
#SVM Classifier
model = SVC(C = 1) # cost C = 1
model.fit(x_train, y_train)
y_pred = model.predict(x_test) # predict
```

```python
kc = metrics.confusion_matrix(y_test, y_pred)
print("SVM accuracy: ", kc)
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier 
from sklearn import metrics

# Generate some sample data for demonstration purposes
X = np.random.rand(100, 5)  # Sample features
y = np.random.randint(0, 2, 100)  # Sample labels (0 or 1)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# K-Nearest Neighbors (KNN) classifier
knn_classifier = KNeighborsClassifier(n_neighbors=3)
knn_classifier.fit(X_train, y_train)
knn_y_pred = knn_classifier.predict(X_test)

# Support Vector Machine (SVM) classifier
svm_classifier = SVC()
svm_classifier.fit(X_train, y_train)
svm_y_pred = svm_classifier.predict(X_test)

# Evaluate performance using accuracy
knn_accuracy = metrics.accuracy_score(y_test, knn_y_pred)
svm_accuracy = metrics.accuracy_score(y_test, svm_y_pred)

# Print the results
print("K-Nearest Neighbors (KNN) Accuracy:", knn_accuracy)
print("Support Vector Machine (SVM) Accuracy:", svm_accuracy)
```
